```python
from haystack.nodes import TextConverter, LinkContentFetcher, PromptNode
from haystack.nodes import OpenAI
from haystack.pipelines import Pipeline

# Initialize components
fetcher = LinkContentFetcher(user_agents=["haystack/LinkContentFetcher/2.0.0b8"], raise_on_failure=True, retry_attempts=2, timeout=3)
converter = TextConverter()  # Use TextConverter instead of HTMLToDocument
llm = OpenAI(api_key=os.environ["OPENAI_API_KEY"], model="gpt-3.5-turbo")
prompt_node = PromptNode(prompt_template="""
According to the contents of this website:
{% for document in documents %}
  {{document.content}}
{% endfor %}
Answer the given question: {{query}}
Answer:
""", llm=llm)

# Build the pipeline
pipeline = Pipeline()
pipeline.add_node(component=fetcher, name="Fetcher", inputs=["Query"])
pipeline.add_node(component=converter, name="Converter", inputs=["Fetcher"])
pipeline.add_node(component=prompt_node, name="PromptNode", inputs=["Converter"])

# Run the pipeline
result = pipeline.run({
    "query": "Which components do I need for a RAG pipeline?",
    "params": {"Fetcher": {"urls": ["https://haystack.deepset.ai/overview/quick-start"]}}
})
print(result["results"][0])
```
